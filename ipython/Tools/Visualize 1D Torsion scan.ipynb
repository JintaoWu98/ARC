# ARC Tools
### Visualize a 1D torsion scan

#### input parameeters:

```python
path = '/path/to/ESS/rotor/scan/output/file/output.out'  # the scan path

folder_path_for_saving = '/path/for/saving/image/'
pivots = [1, 5]  # only used for naming the resulting figure
units = 'degrees'  # either 'degrees' or 'radians'
```

```python
from arc.species.species import determine_rotor_symmetry
from arc.parser import parse_1d_scan_energies
from arc.common import read_yaml_file
from arc.plotter import plot_1d_rotor_scan
%matplotlib notebook
```

```python
if path.endswith('.yml'):
    results = read_yaml_file(path)
    units = 'degrees'
    energies, angles = None, None
else:
    results = None
    energies, angles = parse_1d_scan_energies(path=path)

plot_1d_rotor_scan(angles=angles, energies=energies, results=results,
                   path=folder_path_for_saving, pivots=pivots, units=units)

symmetry, max_e = determine_rotor_symmetry(rotor_path=path, label='label', pivots=pivots)

print(f'\nsymmetry: {symmetry}, maximum barrier hieght: {max_e:.2f} kJ/mol')
```
